### Problema 1

Los siguientes datos representan el costo en dólares de la electricidad según Western Electricity Coordinating Council desde el 2014 (y su pronóstico para el 2040)

Tomado de:
https://www.quandl.com/data/EIA/AEO_2016_REF_NO_CPP_PRCE_NA_COMM_NA_ELC_NA_WECCRKS_NCNTPKWH_A-Electricity-End-Use-Prices-Commercial-Western-Electricity-Coordinating-Council-Rockies-No-CPP-AEO2016

```python
import numpy as np
# Datos extraidos de la página de quandl
data = np.array([10.051101,  9.259127,  9.444813,  9.557723,  9.977324, 10.298326,
       10.447217, 10.550506, 10.590784, 10.657418, 10.743085, 10.831965,
       10.988435, 11.153055, 11.313326, 11.557165, 11.860136, 12.122511,
       12.356448, 12.611391, 12.814569, 13.111372, 13.357134, 13.559012,
       13.85861 , 14.135924, 14.436149])
```

1. Ajuste estos datos a polinomios de grado 1 hasta 3 (los datos en $x$ corresponden a los años de 2014 a 2040, sin embargo son valores muy grandes para elevarlos a una potencia, use una representación para estos años, por ejemplo $2014 \rightarrow 0$ y $2040\rightarrow 26$).

2. Muestre las tres curvas ajustadas junto a los datos de los costos en un solo gráfico

```python
import numpy as np
import matplotlib.pyplot as plt

# Datos de los años (2014 a 2040)
years = np.arange(0, 27)  # Representación de años: 2014 -> 0, 2040 -> 26

# Datos de costo en dólares de la electricidad
data = np.array([10.051101, 9.259127, 9.444813, 9.557723, 9.977324, 10.298326,
                 10.447217, 10.550506, 10.590784, 10.657418, 10.743085, 10.831965,
                 10.988435, 11.153055, 11.313326, 11.557165, 11.860136, 12.122511,
                 12.356448, 12.611391, 12.814569, 13.111372, 13.357134, 13.559012,
                 13.85861, 14.135924, 14.436149])

# Ajuste de polinomios de grado 1 a 3
coeffs_degree1 = np.polyfit(years, data, 1)
coeffs_degree2 = np.polyfit(years, data, 2)
coeffs_degree3 = np.polyfit(years, data, 3)

# Valores estimados para los polinomios
fit_degree1 = np.polyval(coeffs_degree1, years)
fit_degree2 = np.polyval(coeffs_degree2, years)
fit_degree3 = np.polyval(coeffs_degree3, years)

# Crear el gráfico
plt.figure(figsize=(10, 6))
plt.scatter(years, data, label='Datos de Costo de Electricidad', marker='o', color='b')
plt.plot(years, fit_degree1, label='Ajuste Grado 1', color='r')
plt.plot(years, fit_degree2, label='Ajuste Grado 2', color='g')
plt.plot(years, fit_degree3, label='Ajuste Grado 3', color='m')

# Configurar el gráfico
plt.title('Ajuste Polinomial de Costo de Electricidad')
plt.xlabel('Años (2014 → 2040)')
plt.ylabel('Costo en Dólares')
plt.legend(loc='upper left')

# Mostrar el gráfico
plt.grid(True)
plt.show()
```

### Problema 2

El archivo `population_data.csv` contiene información histórica anual de México relativa a la población, nacimientos y muertes desde el año 1950 hasta el año 2015. La primer columna corresponde a los años, la segunda a la población, la tercera a los nacimientos y la última a los decesos.

Tomado de: https://en.wikipedia.org/wiki/Demographics_of_Mexico.

Importe estos datos usando la función `read_csv` del paquete `pandas`

1. Usando los años como variable independiente $x$ y la población como variable dependiente $y$, ajustar polinomios de grado 1 hasta grado 3.

    * Mostrar en un solo gráfico los datos de población contra los años, y los polinomios ajustados.

    * Graficar el error cuadrático acumulado contra el número de términos.

2. Repetir el punto 1 usando los nacimientos como variable dependiente $y$. Grafique las curvas ajustadas junto con los datos de nacimientos. Vea en un gráfico el error acumulado.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Importar los datos desde el archivo CSV
data = pd.read_csv('population_data.csv')

# Extraer los años, población y nacimientos
years = data['Year']
population = data['Poblacion']
births = data['Nacimientos']

# Ajuste de polinomios de grado 1 a 3 para población
coeffs_degree1_pop = np.polyfit(years, population, 1)
coeffs_degree2_pop = np.polyfit(years, population, 2)
coeffs_degree3_pop = np.polyfit(years, population, 3)

# Valores estimados para los polinomios de población
fit_degree1_pop = np.polyval(coeffs_degree1_pop, years)
fit_degree2_pop = np.polyval(coeffs_degree2_pop, years)
fit_degree3_pop = np.polyval(coeffs_degree3_pop, years)

# Ajuste de polinomios de grado 1 a 3 para nacimientos
coeffs_degree1_births = np.polyfit(years, births, 1)
coeffs_degree2_births = np.polyfit(years, births, 2)
coeffs_degree3_births = np.polyfit(years, births, 3)

# Valores estimados para los polinomios de nacimientos
fit_degree1_births = np.polyval(coeffs_degree1_births, years)
fit_degree2_births = np.polyval(coeffs_degree2_births, years)
fit_degree3_births = np.polyval(coeffs_degree3_births, years)

# Calcular el error cuadrático acumulado
error_degree1_pop = np.sum((fit_degree1_pop - population) ** 2)
error_degree2_pop = np.sum((fit_degree2_pop - population) ** 2)
error_degree3_pop = np.sum((fit_degree3_pop - population) ** 2)

error_degree1_births = np.sum((fit_degree1_births - births) ** 2)
error_degree2_births = np.sum((fit_degree2_births - births) ** 2)
error_degree3_births = np.sum((fit_degree3_births - births) ** 2)

# Crear el gráfico para población
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.scatter(years, population, label='Datos de Población', marker='o', color='b')
plt.plot(years, fit_degree1_pop, label='Ajuste Grado 1', color='r')
plt.plot(years, fit_degree2_pop, label='Ajuste Grado 2', color='g')
plt.plot(years, fit_degree3_pop, label='Ajuste Grado 3', color='m')
plt.title('Ajuste Polinomial de Población')
plt.xlabel('Años')
plt.ylabel('Población')
plt.legend(loc='upper left')

# Crear el gráfico para nacimientos
plt.subplot(1, 2, 2)
plt.scatter(years, births, label='Datos de Nacimientos', marker='o', color='b')
plt.plot(years, fit_degree1_births, label='Ajuste Grado 1', color='r')
plt.plot(years, fit_degree2_births, label='Ajuste Grado 2', color='g')
plt.plot(years, fit_degree3_births, label='Ajuste Grado 3', color='m')
plt.title('Ajuste Polinomial de Nacimientos')
plt.xlabel('Años')
plt.ylabel('Nacimientos')
plt.legend(loc='upper left')

# Mostrar el gráfico
plt.tight_layout()
plt.show()

# Gráfico de error cuadrático acumulado
errors_pop = [error_degree1_pop, error_degree2_pop, error_degree3_pop]
errors_births = [error_degree1_births, error_degree2_births, error_degree3_births]

plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.plot(range(1, 4), errors_pop, marker='o')
plt.title('Error Cuadrático Acumulado - Población')
plt.xlabel('Grado del Polinomio')
plt.ylabel('Error Cuadrático Acumulado')

plt.subplot(1, 2, 2)
plt.plot(range(1, 4), errors_births, marker='o')
plt.title('Error Cuadrático Acumulado - Nacimientos')
plt.xlabel('Grado del Polinomio')
plt.ylabel('Error Cuadrático Acumulado')

plt.tight_layout()
plt.show()

```
