## 3. El archivo datosCT.xlsx

```python
import pandas as pd 
data= pd.read_excel("datosCT.xlsx")
data
```

### - Utilidad

```python
data["Utilidad"]=data["Ingreso"]- data["Gasto"]
data
```

### - El promedio de Gastos en los meses que la Utilidad fue negativa

```python
contador= 0
lista=[]
while contador <12:
    if data["Utilidad"][contador]<0:
        lista.append(data["Gasto"][contador])
        contador+=1
    else:
        contador+=1
        
suma = sum(lista)        
cantidad = len(lista)
promedio = suma / cantidad
print(promedio)
```

### - Los ingresos totales del año

```python
contador= 0
lista=[]
while contador <12:
    lista.append(data["Ingreso"][contador])
    contador+=1
    
print(sum(lista))
```
