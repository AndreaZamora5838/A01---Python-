```python
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt
sp.var("x")
```

**Para la siguientes funciones encuentre los puntos críticos, grafique y diga si éste se trata de un máximo o mínimo, local o global.**

$f(x)=1260+117x-9x^2$

```python

def f(x):
    return 1260+117*x-9*(x**2)

df = sp.diff(f(x),x) 
pc = sp.solve(df,x)[0]
pc
```

```python
xn= np.linspace(2,11,201)

plt.figure()
plt.plot(xn, f(xn))
plt.grid()
```

La función tiene un punto crítico en $\frac{13}{2}$
- El punto critico $\frac{13}{2}$ es un maximo global.


$f(x)=5+x+\frac{4}{x}$, para $x>0$

```python

def f(x):
    return 5+x+4/x

df = sp.diff(f(x),x) 
pc = sp.solve(df,x)
pc
```

```python
xn= np.linspace(0,5,501)

plt.figure()
plt.plot(xn, f(xn))
plt.grid()
```

La función tiene un punto crítico cuando x>0 y es 2 
- El punto critico en 2 es un minimo global.

$f(x)=6-2x+\sqrt{18-2x^2}$, para $-3\leq x\leq 3$

```python
def f(x):
    return 6-(2*x)+(18-2*(x**2))**(1/2)

df = sp.diff(f(x),x) 
pc = sp.solve(df,x)
pc
```

```python
xn= np.linspace(-4,4,501)

plt.figure()
plt.plot(xn, f(xn))
plt.grid()
```

La función tiene un punto crítico cuando $-3\leq x\leq 3$ y es -2.44948974278318
- El punto critico en -2.44948974278318 es un maximo global.

$f(x)=\frac{1}{4}x^4+\frac{1}{3}x^3-x^2$ en el intervalo $[-3,1.7]$

```python
def f(x):
    return 6-2*x+(18-2*x**2)**1/2

df = sp.diff(f(x),x) 
pc = sp.solve(df,x)
pc
```
