### $f(x)=1260+117x-9x^2$

```python
def f(x):
    return 1260+117*x-9*(x**2)

df = sp.diff(f(x),x) 
pc = sp.solve(df,x)[0]
pc
```

```python
y = f(pc)
print ("La función tiene un punto crítico en la coordenada (" + str(pc) + ", " + str(y) + ")")
```

```python
xn= np.linspace(2,11,201)

plt.figure()
plt.plot(xn, f(xn))
plt.plot(pc, y, "*b", ms=10)
plt.grid()
```

- El punto crítico ($\frac{13}{2}$, $\frac{6561}{4}$) es un máximo global.

### $f(x)=\frac{1}{4}x^4+\frac{1}{3}x^3-x^2$ en el intervalo $[-3,1.7]$

```python
def f(x):
    return 1/4*x**4+1/3*x**3-x**2

df = sp.diff(f(x),x) 
pc = sp.solve(df,x)
pc
```

```python
for i in pc:
    y = f(i)
    print ("Cuando -3 <= x <= 1.7 la función tiene un punto crítico en la coordenada (" + str(i) + ", " + str(y) + ")")
```

```python
xn= np.linspace(-3,1.7,1000)

plt.figure()
plt.plot(xn, f(xn))
for i in pc:
    y = f(i)
    plt.plot(i, y, "*b", ms=10)
plt.grid()
```

- El punto crítico (-2, $\frac{-8}{3}$) es un mínimo global.
- El punto crítico (0, 0) es un máximo local.
- El punto crítico (1, $\frac{-5}{12}$) es un mínimo local.

### 1. El perímetro de un triángulo isósceles es de $10 cm$. ¿Cuánto deben medir sus lados para que el volumen del cuerpo generado por la rotación del triángulo en torno a su base sea el mayor posible? (Volumen de un cono $= \frac{1}{3}\pi r^2 h$, donde $r$ es el radio de la base y $h$ la altura del cono).

Dado que el triángulo isósceles tiene un perímetro de 10 cm, podemos representar sus lados de la siguiente manera:

- Sea "a" la longitud de los dos lados iguales del triángulo isósceles y "b" la longitud de la base del triángulo.
$$P=2a+b=10cm$$

- Podemos resolver esta ecuación para "b":
$$b = 10 - 2a$$

- Podemos usar el teorema de Pitágoras para expresar la altura en términos de "a" y "b":
$$h^2\:=\:a^2-\left(0.5\:\cdot \:\:b\right)^2\:$$
$$h^2\:=\:a^2-\left(0.5\:\cdot \left(\:10\:-\:2a\right)\:\:\right)^2\:$$
$$h^2\:=\:a^2\:-\:0.25\:\cdot \:\left(10\:-\:2a\right)^2$$
$$h=\sqrt{\:a^2\:-\:0.25\:\cdot \:\:\:\left(10\:-\:2a\right)^2}$$

- El volumen de un cono se calcula utilizando la fórmula: $V\:=\left(\frac{1}{3}\right)\cdot \pi \cdot r^2\cdot h$. En este caso, el radio de la base es $0.5b$ y la altura es $\sqrt{\:a^2\:-\:0.25\:\cdot \:\:\:\left(10\:-\:2a\right)^2}$

$$V=\frac{1}{3}\pi \left(0.5b\right)^2\left(\sqrt{\:a^2\:-\:0.25\:\cdot \:\:\:\left(10\:-\:2a\right)^2}\right)$$
$$V=\frac{1}{3}\pi \left(0.5\cdot \left(10\:-\:2a\right)\right)^2\:\:\left(\sqrt{\:a^2\:-\:0.25\:\cdot \left(10\:-\:2a\right)^2}\right)$$
$$V\:=\:\frac{1}{2}\pi \left(10\:-\:2a\right)^2\:\left(\sqrt{\:a^2\:-\:0.25\:\cdot \left(10\:-\:2a\right)^2}\right)$$

```python
import sympy as sp

#Derivar e igualar a cero para encontrar los puntos críticos

sp.var("a")
def V(a):
    return 1/2*sp.pi*(10 - 2*a)**2 * ((a**2 - 0.25 * (10 - 2*a)**2))**(1/2)
df = sp.diff(V(a), a)
df
```

```python
pc = sp.solve(df,a)
pc
```

```python
#Criterio de la segunda derivada
ddf= sp.diff(df,a)
for i in pc:
    crit_2da_derivada = ddf.subs(a, i)
    if crit_2da_derivada < 0:
        print("El punto crítico se encuentra en "+ str(i) + " y es un máximo")
    elif crit_2da_derivada > 0:
        print("El punto crítico se encuentra en "+ str(i) + " y es un mínimo")
    else:
        print("El punto crítico se encuentra en "+ str(i) + " y con este método no se puede saber que es")
```

De esta forma ya sabemos que 3 es el valor para a donde sera el máximo de volumen.
Por lo tanto, ya podemos saber cunto mide b.

```python
b= 10-2*3
b
```

Respueta:
Por lo tanto, para maximizar el volumen del cono, los lados iguales del triángulo isósceles deben medir $3 cm$ y la base del triángulo debe medir $4 cm$.

### 2. Disponemos de una barra de aluminio de 6 metros para construir una portería de fútbol. Si queremos que el área de la portería sea máxima, ¿cuánto deben medir los postes y el travesaño?

$x$ es la longitud de cada uno de los postes verticales, y $y$ es la longitud del travesaño horizontal. La ecuación que representa la longitud total del trozo de barra de aluminio utilizado para construir la portería es:
$$2x + y = 6$$
Para expresar el área de la portería en términos de "x" y "y", debemos calcular el producto de la longitud del travesaño "y" y la longitud de uno de los postes "x".
$$A=y\cdot x$$
Ahora, podemos expresar "y" en términos de "x"y sustituimos esta expresión en la fórmula del área:
$$y=6-2x$$

$$A=\left(6-2x\right)x=6x-2x^2$$

```python
#Derivar e igualar a cero para encontrar los puntos críticos
sp.var("x")
def A(x):
    return 6*x - 2*x**2
df = sp.diff(A(x), x)
df
```

```python
pc = sp.solve(df,x)[0]
pc
```

```python
#Graficar para berificar que es un punto máximo
y = A(pc)
print ("La función tiene un punto crítico en la coordenada (" + str(pc) + ", " + str(y) + ")")

xn= np.linspace(-1,4,100)
plt.figure()
plt.plot(xn, A(xn))
plt.plot(pc, y, "*b", ms=10)
plt.grid()
```

Ahora que tenemos el valor de "x", podemos encontrar el valor correspondiente de "y" usando la ecuación de la longitud total de la barra:

```python
y = 6 - 2 * 1.5
y 
```

Respuesta:
Por lo tanto, para maximizar el área de la portería de fútbol, los postes deben medir 1.5 metros de longitud y el travesaño debe medir 3 metros de longitud.
